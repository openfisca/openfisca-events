```python
#Specific usage imports
import openfisca_france
```

Initialize the tax and benefit system of France

```python
tax_benefit_system = openfisca_france.FranceTaxBenefitSystem()
```

Create a scenario

```python
def make_single_with_child_scenario( year = None, tax_benefit_system = tax_benefit_system):
    scenario = tax_benefit_system.new_scenario().init_single_entity(
        period = year,
        parent1 = dict(
            salaire_de_base = 100000,
            age = 40,
            ),
        )
    return scenario
```

```python
simulation = make_single_with_child_scenario(year = 2005).new_simulation()
simulation.calculate('impo')
```

```python
simulation = make_single_with_child_scenario(year = 2011).new_simulation()
```

```python
simulation.calculate('impo')
```

```python
broken_years = list()
for year in range(2002,2015):
    try:
        simulation = make_single_with_child_scenario(year = year).new_simulation()
        simulation.calculate('impo')
    except:
        print sys.exc_info()[0]
        broken_years.append(year)
```
