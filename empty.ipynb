```python
import collections
import json
import urllib2
```

```python
simulation = dict(
    scenarios = [
        dict(
            test_case = dict(
                familles = [
                    dict(
                        parents = ['ind0'],
                        ),
                    ],
                foyers_fiscaux = [],
                individus = [
                    dict(
                        activite = u'Retraité',
                        birth = '1940-01-01',
                        id = 'ind0',
                        rsti = 12500,
                        statmarit = u'Marié',
                        ),
                    ],
                menages = [],
                ),
            year = 2013,
            ),
        ],
    validate = True,
    )
```

```python
request = urllib2.Request('http://localhost:2014/api/1/simulate', headers = {
    'Content-Type': 'application/json',
    'User-Agent': 'OpenFisca-Notebook',
    })
try:
    response = urllib2.urlopen(request, json.dumps(simulation))
except urllib2.HTTPError as response:
    print response.read()
    raise
response_dict = json.loads(response.read(), object_pairs_hook = collections.OrderedDict)
print json.dumps(response_dict, indent = 2)
variables = response_dict['value']
```

```python
for variable in variables:
    print u'{:40s} | {}'.format(variable['code'], ' | '.join('{:6.0f}'.format(value) for value in variable['values']))
```
